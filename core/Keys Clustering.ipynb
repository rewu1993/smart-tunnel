```python
def clean_key(k):
    clean_words = []
    for w in word_tokenize(k):
        if len(w)>1 and not w.isdigit():
            for w_s in w.split('-'):
                if not "." in w_s:
                    clean_words.append(w_s)
    return clean_words
           
```
